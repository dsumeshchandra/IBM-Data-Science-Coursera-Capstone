```python
import requests
```

```python
post_codes_ca_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

## Webscrapping data from Wikipedia

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(post_codes_ca_url,'lxml')
```

## Extratct data into list codes, borough and neighborhood

```python
codes=[]
borough=[]
neighborhood=[]
i=1
for tag in soup.table.find_all('td'):
    if i == 1:
        codes.append(tag.text)
    if i == 2:
        borough.append(tag.text)
    if i == 3: 
        neighborhood.append(tag.text)
    i = i+1
    if i==4:
        i=1
```

## Merging Lists into dataframe

```python
import pandas as pd
import numpy as np

toronto_df = pd.DataFrame(columns=['Postalcode','Borough','Neighborhood'])
toronto_df['Postalcode'] = codes
toronto_df['Borough'] = borough
toronto_df['Neighborhood'] = neighborhood
toronto_df
```

## Identify indexes of Borough "Not Assigned"

```python
indexs = toronto_df[toronto_df['Borough']=='Not assigned'].index
indexs
```

## remove the rows with 'Not assigned' Borough

```python
toronto_df.drop(toronto_df.index[indexs], inplace=True)
```

## remove the '\n' in Neighborhood at all rows
## reset index

```python
toronto_df['Neighborhood'] = toronto_df.Neighborhood.str.replace("\n","")
toronto_df = toronto_df.reset_index(drop=True)
toronto_df
```

## Get the indexs of which Neighborhood are 'Not assigned'

```python
indexs_neighborhood = toronto_df[toronto_df['Neighborhood']=="Not assigned"].index
indexs_neighborhood
```

## Assigning the Neighbourhood based on values from Borough

```python
for i in range(210):
    if toronto_df.Neighborhood[i]=='Not assigned':
        toronto_df.Neighborhood[i] = toronto_df.Borough[i]
```
